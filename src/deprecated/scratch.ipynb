```python
def make_env():
    env = gym.make('ALE/Breakout-v5', frameskip=4, repeat_action_probability=0.0, full_action_space=False)  # Use equivalent parameters to BreakoutDeterministic-v4
    return env

num_envs = 3
envs = SyncVectorEnv([make_env for _ in range(num_envs)], autoreset_mode=gym.vector.AutoresetMode.NEXT_STEP)
states, _ = envs.reset()
print("State Means after Reset: ", [np.mean(state) for state in states])
actions = [1] * num_envs
next_states, rewards_batch, terminated, truncated, infos = envs.step(actions)
for i in range(1000):
    actions = [random.choice([0,1,2,3]) for _ in range(num_envs)]
    # print("Actions: ", actions)
    next_states, rewards_batch, terminated, truncated, infos = envs.step(actions)
print("Next States Means: ", [np.mean(state) for state in next_states])
print("Rewards Batch: ", rewards_batch)
print("Terminated: ", terminated)
print("Truncated: ", truncated)
print("Infos: ", infos)
print(type(terminated))

```
